```python
import numpy as np
import matplotlib.pyplot as plt


```

```python
X = np.arange(10)
Y = (X-5)**2 + 3
print(X,Y)
```

```python
# visualize
plt.style.use("seaborn")
plt.plot(X,Y)
plt.show()
```

```python
# Gradient descent
# init x with any random value
x = 0
lr = 0.1
plt.plot(X,Y)
for i in range(50):
  grad = 2*(x-5)
  x = x - lr * grad
  y = (x-5)**2 + 3
  plt.scatter(x,y)

plt.show()


```
